```python
%matplotlib inline

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

import random
from pprint import pprint

set_cnt = 10
N = 10
x = [np.random.rand(N) for i in range(set_cnt)]
y = [np.random.rand(N) for i in range(set_cnt)]
area = [(30 * np.random.rand(N)) ** 2 for i in range(set_cnt)]
t = [i / set_cnt for i in range(set_cnt)]

fig, ax = plt.subplots()

cmap = plt.get_cmap('hsv')

for i in range(set_cnt):
    ax.scatter(x[i], y[i], s=area[i], c = cmap(t[i]))

# call plt.show() if you want to create the plot window
# plt.show() 
```
